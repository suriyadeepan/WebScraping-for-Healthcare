# COVID19 API

```python
import requests
import json
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

## Live Data from India

```python
resp = requests.get('https://api.covid19api.com/live/country/india')
data = json.loads(resp.content)
df = pd.DataFrame(data)
df
```

```python
df.Province.unique()
```

```python
df[df.Province == 'Puducherry']
```

```python
import matplotlib

df["DateFormatted"] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
dates = matplotlib.dates.date2num(df[df.Province == 'Puducherry'].DateFormatted.values)
plt.figure(figsize=(12, 8))
plt.plot_date(dates, df[df.PDeathsrovince == 'Puducherry'].Deaths.values, linestyle='-', markersize=4, c='red', alpha=0.4);
```

## From January to October

```python
_from = "2021-01-01T00:00:00Z"
_to = "2021-10-20T00:00:00Z"
country = 'India'.lower()
resp = requests.get(
    f'https://api.covid19api.com/country/{country}/status/confirmed?from={_from}&to={_to}'
)
data = json.loads(resp.content)
df = pd.DataFrame(data)
```

```python
df
```

```python
dates = matplotlib.dates.date2num(df.Date)
plt.figure(figsize=(15, 10))
plt.plot_date(dates, df.Cases.values, linestyle='-', markersize=4, c='red', alpha=0.4);
```

## International Summary

```python
summary = json.loads(requests.get("https://api.covid19api.com/summary").content)
summary
```

```python
countries = pd.DataFrame(summary['Countries'])
countries
```

```python
countries_sorted = countries.sort_values(axis=0, by="TotalDeaths", ascending=False).reset_index(drop=True)
```

```python
plt.figure(figsize=(8, 8))
plt.barh(countries_sorted.Country.loc[:10], countries_sorted.TotalDeaths.loc[:10]);
plt.gca().invert_yaxis()
```
